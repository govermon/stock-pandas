```python
import pandas as pd
from stock_pandas import StockDataFrame
```

```python
stock = StockDataFrame(pd.read_csv('tencent.csv'), date_col='time_key')

display(stock)
stock.info()
```

```python
stock['ma:5']
```

```python
import mplfinance as mpf
```

```python
stock.alias('Open', 'open')
stock.alias('High', 'high')
stock.alias('Low', 'low')
stock.alias('Close', 'close')
```

```python
apds = [
    mpf.make_addplot(stock[['ma:5', 'ma:10', 'ma:30']])
]

mpf.plot(stock, type='candle', addplot=apds, figscale=2)
```
